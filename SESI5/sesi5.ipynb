```python
fastaFile = SeqIO.read("sequence.gb", "genbank")
seqMain = fastaFile.seq
print(seqMain)
```

```python
length = round(len(seqMain) / 2)
seqA = seqMain[:length]
seqB = seqMain[length:]
```

```python
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp
import matplotlib.pyplot as plt
```

```python
gcA = GC(seqA)
gcB = GC(seqB)

mtA = MeltingTemp.Tm_NN(seqA)
mtB = MeltingTemp.Tm_NN(seqB)
```

```python
_, axs = plt.subplots(1,2)
axs[0].bar(["seqA", "seqB"], [gcA, gcB])
axs[0].set_title("GC content")
axs[1].bar(["seqA", "seqB"], [mtA, mtB])
axs[1].set_title("Melting temperature")
plt.show()
```
