AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC
# Sequence B:
CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG


Please use Biopython to:
•	Determine the molecular weight of both sequence.
•	Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library.
•	Calculate the GC content of sequence A and sequence B.
•	Calculate the AT content of sequence A and sequence B.
•	Plot the GC and AT content of sequence A and sequence B using Matplotlib library.


```python
from Bio.Seq import Seq
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import molecular_weight
import matplotlib.pyplot as plt
```

```python
Sequence_A = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
Sequence_B = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')

print("Sequence A : ", Sequence_A)
print("Sequence B : ", Sequence_B)
print("1. Determine the molecular weight of both sequences")
weight_A = molecular_weight(Sequence_A)
weight_B = molecular_weight(Sequence_B)
print("Molecular weight of Sequence A : ", weight_A)
print("Molecular weight of Sequence B : ", weight_B)
```

```python
print("2. Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library")
mt_A = mt.Tm_Wallace(Sequence_A)
mt_B = mt.Tm_Wallace(Sequence_B)
print("Melting point of Sequence A : ", mt_A, "°C")
print("Melting point of Sequence B : ", mt_B, "°C")

print("Plotting")
color = ["Blue", "Red"]
plt.bar(["Wallace A", "Wallace B"], [mt_A, mt_B],color=color)
plt.xlabel("Sequences")
plt.ylabel("Melting Point")
plt.title("Melting Point of Sequences")
plt.show()
```

```python
print("3. Calculate the GC content of sequence A and sequence B")

gc_A = GC(Sequence_A)
gc_B = GC(Sequence_B)
print("GC content of Sequence A : ", gc_A,"%")
print("GC content of Sequence B : ", gc_B,"%")
```

```python
print("4. Calculate the AT content of sequence A and sequence B")
at_A = 100 - GC(Sequence_A)
at_B = 100 - GC(Sequence_B)
print("AT content of Sequence A : ", at_A,"%")
print("AT content of Sequence B : ", at_B,"%")
```

```python
print("5. Plot the GC and AT content of sequence A and sequence B using Matplotlib library.")
plt.bar(["Sequence A", "Sequence B"], [gc_A, gc_B],color=color)
plt.xlabel("Sequences")
plt.ylabel("GC Content")
plt.title("GC Content of Sequences")
plt.show()

plt.bar(["Sequence A", "Sequence B"], [at_A, at_B],color=["green","yellow"])
plt.xlabel("Sequences")
plt.ylabel("AT Content")
plt.title("AT Content of Sequences")
plt.show()
```
