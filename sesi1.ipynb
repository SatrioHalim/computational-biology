```python
from Bio.Seq import Seq

seq_A = Seq("ACGTCGTAACGGTA")

print(seq_A)

seq_B = Seq("ACGTGCATGCTAGTCATGCT")

join_sequence = seq_A + seq_B
print(join_sequence)

#Length, cari tau panjang sequences
len_A = len(seq_A)
len_B = len(seq_B)

print(len_A, len_B)

# Count
total_A = seq_A.count("A")
total_C = seq_A.count("C")
total_G = seq_A.count("G")
total_T = seq_A.count("T")

print(total_A, total_C, total_G, total_T)
```

```python
# Find & rfind & index
# Find ("String", start)
# ACGTCGTAACGGTA

find_A = seq_A.find("T",4)
print(find_A)

index_B = seq_A.index("G")
print(index_B)

rfind_C = seq_A.rfind("G")
print(rfind_C)
```

```python
# nt_Search, cari subsequence pada sequence
# nt_search typecast sequence menjadi string, dgn str()

from Bio.SeqUtils import nt_search

seq_baru = Seq("ACGTCGTAACGGTA")
sub_seq = Seq("CGT")

print(nt_search(str(seq_baru), str(sub_seq)))
```

```python
# Slicing & Reverse
# Slicing -> arr[Start:End(excluded):step]
# ACGTCGTAACGGTA -> seq_baru

print("tanpa penggunaan step", seq_baru[2:5])
print("dengan penggunaan step", seq_baru[2:8:2])
print("Reverse seq_baru", seq_baru[::-1])
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from Bio.SeqUtils import nt_search
from Bio.Seq import Seq

sequence_A = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTCG")
sequence_B = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")

# •	Find the length of both sequences.
print("Find the length of both sequences : ")
length_A = len(sequence_A)
length_B = len(sequence_B)
print("Length of sequence A:", length_A)
print("Length of sequence B:", length_B)
print("")

#•	Determine the number of times the codon (CGC) appears in both sequences

sub_sequence2 = Seq("CGC")

print("Number of times the codon (CGC) appears in both sequences : ")
print("Sequence A:", nt_search(str(sequence_A), str(sub_sequence2)))
print("Sequence B:", nt_search(str(sequence_B), str(sub_sequence2)))
print("")

# •	Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences

sub_sequence3 = Seq("CAGTC")
print("Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences : ")
print("Sequence A:", sequence_A.find(sub_sequence3))
print("Sequence B:", sequence_B.find(sub_sequence3))
print("")

# •	Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a 
#   new DNA sequence called sequence C.
print("Combine the first 15 nucleotides from sequence A and the last 10 nucleot ides from sequence B to create a new DNA sequence called sequence C : ")
sequence_C = sequence_A[:15] + sequence_B[-10:]
print(sequence_C)
print("")

#•	Reverse  the sequence of sequence C 
print("Reverse the sequence of sequence C : ")
sequence_C_reverse = sequence_C[::-1]
print(sequence_C_reverse)
print("")

# •	Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.
print("Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library : ")
freq_A = Counter(sequence_A)
freq_B = Counter(sequence_B)
freq_C = Counter(sequence_C)
color = ["Blue", "Red", "Green", "Yellow"]
plt.title('Nucleotide base frequency of sequence A')
plt.bar(freq_A.keys(), freq_A.values(),color=color)
plt.xlabel('Nucleotide Bases')
plt.ylabel('Frequency')
plt.show()

plt.title('Nucleotide base frequency of sequence B')
plt.bar(freq_B.keys(), freq_B.values(),color=color)
plt.xlabel('Nucleotide Bases')
plt.ylabel('Frequency')
plt.show()

plt.title('Nucleotide base frequency of sequence C')
plt.bar(freq_C.keys(), freq_C.values(),color=color)
plt.xlabel('Nucleotide Bases')
plt.ylabel('Frequency')
plt.show()

```
